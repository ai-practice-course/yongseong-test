## naver_kospi_crawler (코스피 지수)

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import openpyxl

tr_dates = []
tr_kospis = []
new_tr_kospis = []

target_dates = []
target_kospis = []

for n in range(1, 4):
  res = requests.get(f'https://finance.naver.com/sise/sise_index_day.naver?code=KOSPI&page={n}')
  soup = BeautifulSoup(res.content, 'html.parser')

  start_date = '2022.03.30'
  end_date = '2022.03.11'

  trs = soup.find_all('tr')
  for tr in trs:
    if tr.find('td', 'date'):
      dates = tr.find('td', 'date')
      for date in dates:
        if end_date < date < start_date:
          tr_dates.append(date)
          tr_dates = list(set(tr_dates))
          tr_dates.sort(reverse=True)
          kospis = tr.find('td', 'number_1')
          for kospi in kospis:
            tr_kospis.append(kospi)
          for tr_kospi in tr_kospis:
            tr_kospi = tr_kospi.replace(",","")
            tr_kospi = float(tr_kospi)
            if tr_kospi not in new_tr_kospis:
              new_tr_kospis.append(tr_kospi)

target_dates.extend(tr_dates)    
target_kospis.extend(new_tr_kospis) 

print(target_dates)  # 삭제해도 무방함
print(target_kospis)  # 삭제해도 무방함

zip_list = zip(target_dates, target_kospis)
item_list = list(zip_list)

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

for item in item_list:
  excel_sheet.append(list(item))

excel_file.save('kospis.xlsx')
excel_file.close
```

## naver_exchanges_crawler (달러 환율 지수)

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import openpyxl

tr_dates = []
tr_exchanges = []

target_dates = []
target_exchanges = []

for n in range(1, 4):
  res = requests.get(f'https://finance.naver.com/marketindex/exchangeDailyQuote.naver?marketindexCd=FX_USDKRW&page={n}')
  soup = BeautifulSoup(res.content, 'html.parser')

  start_date = '2022.04.05'
  end_date = '2022.02.21'

  trs = soup.find_all('tr')

  for tr in trs:
    if tr.find('td', 'date'):
      dates = tr.find('td', 'date')
      date = dates.get_text()
      if end_date < date < start_date:
        tr_dates.append(date)
        exchange = tr.find('td', 'num').get_text()
        exchange = exchange.replace(",","")
        exchange = float(exchange)
        tr_exchanges.append(exchange)

target_dates.extend(tr_dates)    
target_exchanges.extend(tr_exchanges) 

print(target_dates)
print(target_exchanges)

zip_list = zip(target_dates, target_exchanges)
item_list = list(zip_list)

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

for item in item_list:
  excel_sheet.append(list(item))

excel_file.save('exchanges.xlsx')
excel_file.close
```

## naver_rates_crawler (국고채(3년) 금리 지수)

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import openpyxl

tr_dates = []
tr_rates = []

target_dates = []
target_rates = []

for n in range(1, 5):
  res = requests.get(f'https://finance.naver.com/marketindex/interestDailyQuote.naver?marketindexCd=IRR_GOVT03Y&page={n}')
  soup = BeautifulSoup(res.content, 'html.parser')

  start_date = '2022.04.05'
  end_date = '2022.02.23'

  trs = soup.find_all('tr')

  for tr in trs:
    if tr.find('td', 'date'):
      dates = tr.find('td', 'date')
      date = dates.get_text()
      date = date.strip()
      if end_date < date < start_date:
        tr_dates.append(date)
        rate = tr.find('td', 'num').get_text()
        rate = float(rate)
        tr_rates.append(rate)

target_dates.extend(tr_dates)    
target_rates.extend(tr_rates) 

print(target_dates)
print(target_rates)

zip_list = zip(target_dates, target_rates)
item_list = list(zip_list)

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

for item in item_list:
  excel_sheet.append(list(item))

excel_file.save('rates.xlsx')
excel_file.close
```
